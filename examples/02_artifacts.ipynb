# qbalance Artifact Notebook
Persist run artifacts and validate outputs.

```python
import logging
from pathlib import Path
from tempfile import TemporaryDirectory

from qbalance import Workload, load_data

logging.disable(logging.ERROR)
```

```python
tmp = TemporaryDirectory()
out = Path(tmp.name)
ds = load_data("tiny")
wl = Workload.from_dataset(ds).set_target("fake:generic:5")
balanced = wl.adjust(search="grid", pareto=False, max_candidates=8, execute=False, profile=False, seed=11)
run_dir = out / "balanced"
zip_path = out / "balanced.zip"
balanced.save(run_dir, overwrite=True)
balanced.to_download(zip_path, overwrite=True)
print(run_dir.exists(), zip_path.exists())
```
