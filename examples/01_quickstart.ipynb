# qbalance Quickstart Notebook
Run a minimal balancing flow and inspect summary metrics.

```python
import logging

from qbalance import Workload, load_data

logging.disable(logging.ERROR)

ds = load_data("tiny")
wl = Workload.from_dataset(ds).set_target("fake:generic:5")
balanced = wl.adjust(search="bandit", pareto=True, max_candidates=16, execute=False, profile=False, seed=3)
print(balanced.summary())
```
